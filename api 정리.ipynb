````python
import requests
from IPython.display import Image, display

url = "https://api.thecatapi.com/v1/images/search"
response = requests.get(url)
data = response.json()
cat_url = data[0]['url']

print("오늘의 고양이 사진:")
display(Image(url=cat_url))
````

날씨 api

> 주소 open meteo api

https://open-meteo.com/

>기능

 온습도 상대 습도 기상 코드 해수면 기압 풍속 풍향 기온 토양 온도 및 수분 등 다양한 시간별 기능이 있다.

>장점

- 오픈소스이며 라이선스를 받아 상업적 목적으로 데이터를 자유롭게 공유하고 조정이 가능하다
- HTTP 프로토콜을 채택하고 JSON 데이터 형식이기 때문에 사용하기에 편리하다
- 80년간의 역사 자료를 가지고 있다.


````python
import requests
from datetime import datetime
latitude = 37.5665
longitude = 126.9780
today = datetime.utcnow().date()
#today= '2025-04-14'
url = (
    f"https://api.open-meteo.com/v1/forecast?"
    f"latitude={latitude}&longitude={longitude}"
    f"&daily=temperature_2m_max,temperature_2m_min&timezone=Asia%2FSeoul"
)
response = requests.get(url)
data = response.json()
dates = data['daily']['time']
max_temps = data['daily']['temperature_2m_max']
min_temps = data['daily']['temperature_2m_min']

for date in dates:
    if date == str(today):
        i = dates.index(date)
        print(f"오늘({today}) 서울 날씨")
        print(f"- 최고 기온: {max_temps[i]}°C")
        print(f"- 최저 기온: {min_temps[i]}°C")
        break
````

> 호출의 기본 요소

https://zenquotes.io/api/[mode]/[key]?option1=value&option2=value

https://zenquotes.io/api = ZenQuotes API URL

[모드] = 검색 유형 [인용, 오늘, 저자, 무작위] 필수

[키] = 프리미엄 구독용 API 키, 하이재킹을 방지하기 위해 소스 코드에서 반드시 
이를 숨기거나 난독화하세요. 선택 사항

[옵션] = 추가 옵션. 선택 사항.

q = 인용 텍스트

a = 저자 이름

I = 저자 이미지 (키 필요)

c = 문자 수

h = 사전 형식화된 HTML 인용

````python
import requests

url = "https://zenquotes.io/api/random"
response = requests.get(url)
data = response.json()
print(data)
import json
print(json.dumps(data, indent=3))

quote = data[0]['q']
author = data[0]['a']

print("오늘의 명언:")
print(f'"{quote}" — {author}')
````
