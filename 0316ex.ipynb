```python
cof = int(input('몇 잔 주문하시겠습니까. 1잔에 300원입니다.'))
mon = int(input('얼마를 넣으시겠습니까.'))

while True :
    if mon == 300 :
        print ('커피' + cof + '잔을 줍니다.')
        cof = cof - 1 
    elif mon > 300 :
        print ('커피를 줍니다. 남은 커피는 %d개입니다' %cof)
        cof = cof -1
        print ('남은 돈은 %d 입니다' % (mon - 300*cof))
           
        
       
    else :
        print ('돈을 다시 돌려주고 커피를 주지 않습니다.')
        break

    if cof == 0:
        print ('커피가 다 떨어졌습니다.')
        break
```
